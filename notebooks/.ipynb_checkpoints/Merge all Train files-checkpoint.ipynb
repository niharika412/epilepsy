```python
from sklearn import model_selection
```

```python
train['kfold']=-1
train = train.sample(frac=1).reset_index(drop=True)
kf = model_selection.KFold(n_splits=5)
for fold, (trn_, val_) in enumerate(kf.split(X=train)):
    train.loc[val_, 'kfold'] = fold
```
